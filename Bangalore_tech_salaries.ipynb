```
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
```

```
print(df.columns.tolist())
```

```
df['company_type_clean'] = df['company_type'].str.strip().str.title()

print(df['company_type_clean'].value_counts())
```

```
df['education_tier_clean'] = df['education_tier'].replace({
    'T1':'Tier 1',
    '1':'Tier 1',
    'Tier-1':'Tier 1',
    'T2':'Tier 2',
    '2':'Tier 2',
    'Tier-2':'Tier 2',
    'T3':'Tier 3',
    '3':'Tier 3',
    'Tier-3':'Tier 3'
})

df['education_tier_clean'] = df['education_tier_clean'].fillna(df['education_tier'])

print(df['education_tier_clean'].value_counts())
```

```
def convert_ctc(ctc):

    if pd.isna(ctc):
        return np.nan

    ctc = str(ctc).replace('₹','').replace('LPA','').replace(',','').strip()

    value = float(ctc)

    if value > 1000:
        value = value/100000

    return value
```

```
df['current_ctc_lpa'] = df['current_ctc'].apply(convert_ctc)

df['previous_ctc_lpa'] = df['previous_ctc'].apply(convert_ctc)

df[['current_ctc','current_ctc_lpa']].head()
```

```
df = df.drop_duplicates()

print("Dataset Shape:", df.shape)
```

```
print(df.dtypes)
```

```
df['role_clean'] = df['role_'].replace({
    'Backend Engineer': 'SDE Backend',
    'Backend Dev': 'SDE Backend',
    'Backend Developer': 'SDE Backend',
    'BE': 'SDE Backend',
    'Frontend Engineer': 'SDE Frontend',
    'Frontend Dev': 'SDE Frontend',
    'Full Stack Developer': 'SDE Full Stack',
    'Full Stack Engineer': 'SDE Full Stack',
    'SDE Full-Stack': 'SDE Full Stack'
})

df['role_clean'] = df['role_clean'].fillna(df['role_'])
```

TASK 3.1

```
role_salary = df.groupby('role_clean')['current_ctc_lpa'].agg(
    Median_CTC='median',
    Mean_CTC='mean',
    Min_CTC='min',
    Max_CTC='max'
).sort_values(by='Median_CTC', ascending=False)

print(role_salary.round(2))
```

TASK 3.2

```
backend = df[df['role_clean'] == 'SDE Backend'].copy()

bins = [0, 2, 4, 6, 100]

labels = ['0-1', '2-3', '4-5', '6+']

backend['exp_band'] = pd.cut(
    backend['years_exp'],
    bins=bins,
    labels=labels,
    right=False
)

exp_salary = backend.groupby('exp_band')['current_ctc_lpa'].median()

print(exp_salary.round(2))
```

TASK 3.3

```
sde = df[df['role_clean'].isin([
    'SDE Backend',
    'SDE Frontend',
    'SDE Full Stack'
])]

skills = ['AWS', 'ML', 'System Design', 'Kubernetes']

for skill in skills:

    with_skill = sde[sde['skills'].str.contains(skill, case=False, na=False)]

    without_skill = sde[~sde['skills'].str.contains(skill, case=False, na=False)]

    median_with = with_skill['current_ctc_lpa'].median()

    median_without = without_skill['current_ctc_lpa'].median()

    premium = ((median_with - median_without) / median_without) * 100

    print(f"{skill}")
    print(f"With Skill : {median_with:.2f} LPA")
    print(f"Without Skill : {median_without:.2f} LPA")
    print(f"Premium : {premium:.2f}%")
    print("-"*40)
```

TASK 3.4

```
backend = df[df['role_clean'] == 'SDE Backend']

company_salary = backend.groupby('company_type_clean')['current_ctc_lpa'].median().sort_values(ascending=False)

print(company_salary.round(2))

if 'Unicorn' in company_salary.index and 'MNC' in company_salary.index:
    unicorn = company_salary['Unicorn']
    mnc = company_salary['MNC']

    premium = ((unicorn - mnc) / mnc) * 100

    print(f"\nUnicorn Premium over MNC = {premium:.2f}%")
```

TASK 3.5

```
bins = [0,2,4,6,100]

labels = ['0-1','2-3','4-5','6+']

df['exp_band'] = pd.cut(df['years_exp'],
                        bins=bins,
                        labels=labels,
                        right=False)


group_size = df.groupby(
    ['role_clean','company_type_clean','exp_band']
)['current_ctc_lpa'].transform('count')


group_median = df.groupby(
    ['role_clean','company_type_clean','exp_band']
)['current_ctc_lpa'].transform('median')


df['group_size'] = group_size

df['group_median'] = group_median

df['gap'] = df['current_ctc_lpa'] - df['group_median']



underpaid = df[df['group_size'] >= 10]



top10 = underpaid.sort_values('gap').head(10)

print(top10[['employee_id',
             'role_clean',
             'company_type_clean',
             'years_exp',
             'current_ctc_lpa',
             'group_median',
             'gap']])
```
